# Método de Derivada de Segunda ordem - LoG

```python
import cv2
import numpy as np
```

```python
# Carregando a imagem
image = cv2.imread('rim.jpg', cv2.IMREAD_GRAYSCALE)

# Aplicando o filtro Laplaciano do Gaussiano (LoG)
log_image = cv2.GaussianBlur(image, (5, 5), 0)  # Aplica o filtro Gaussiano
log_image = cv2.Laplacian(log_image, cv2.CV_64F)  # Aplica o filtro Laplaciano

# Convertendo a imagem para valores não negativos
log_image = np.abs(log_image)

# Normalizando os valores de pixel para o intervalo 0-255
log_image = cv2.normalize(log_image, None, 0, 255, cv2.NORM_MINMAX)

# Convertendo para o tipo de dado uint8
log_image = np.uint8(log_image)

# Exibindo a imagem original e a imagem com bordas detectadas
cv2.imshow('Imagem Original', image)
cv2.imshow('Bordas Detectadas (LoG)', log_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Filtros de derivada de primeira ordem

```python
import skimage
from skimage import io
import matplotlib
from matplotlib import pyplot as plt
```

## Roberts

```python
img = cv2.imread('rim.jpg', cv2.IMREAD_GRAYSCALE)


plt.imshow(img, cmap='gray')
plt.show()

edges = skimage.filters.roberts(img)

plt.imshow(edges, cmap='gray')
plt.show()
```

## Prewitt

```python
plt.imshow(img, cmap='gray')
plt.show()

edges2 = skimage.filters.prewitt(img)

plt.imshow(edges2, cmap='gray')
plt.show()
```

## Sobel

```python
plt.imshow(img, cmap='gray')
plt.show()

edges3 = skimage.filters.sobel(img)

plt.imshow(edges3, cmap='gray')
plt.show()
```
